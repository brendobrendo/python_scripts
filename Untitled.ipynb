```python
import matplotlib.pyplot as plt
import numpy as np

# Set the desired total profit
total_profit = 1750

# Create a range of possible profit margins
profit_margins = np.arange(1, 11, 1)

# Calculate the number of pizzas to be sold at each profit margin
pizzas_to_sell = total_profit / profit_margins

# Create the plot
plt.figure(figsize=(10,6))
plt.plot(profit_margins, pizzas_to_sell, marker='o')
plt.title('Number of Pizzas to Sell for a $1,750 Profit at Different Profit Margins')
plt.xlabel('Profit Margin ($ per Pizza)')
plt.ylabel('Number of Pizzas to Sell')
plt.grid(True)
plt.show()
```
